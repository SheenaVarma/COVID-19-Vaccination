### Heatmap

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the columns defined above.
corr = sns.heatmap(df.corr(), annot = True, ax = ax) 
```

There is a medium to strong relationship between total_vaccinations and people_fully_vaccinated. However, this visualization is not required for my analysis as I need to determine which countries are progressing with vaccinations.

### Scatterplot

```python
# Create a scatterplot for the "total_vaccinations" and "people_fully_vaccinated_per_hundred" columns in seaborn

sns.lmplot(x = 'total_vaccinations', y = 'people_fully_vaccinated_per_hundred', data = df)
```

There is a positive correlation between total_vaccinations and people_fully_vaccinated_per_hundred but there are many points that are outside the line and this could be due to the "0" value I had to impute in the missing fields during the dataset cleanup.

### Categorical Plots

```python
# Use a histogram to visualize the distribution of the dependent variable - people_fully_vaccinated_per_hundred
sns.histplot(df['people_fully_vaccinated_per_hundred'], bins = 5, kde = True)
```

```python
df.loc[df['people_fully_vaccinated_per_hundred'] < 30, 'Vax category'] = 'Low vaccination rates'
```

```python
df.loc[(df['people_fully_vaccinated_per_hundred'] >= 30) & (df['people_fully_vaccinated_per_hundred'] < 70), 'Vax category'] = 'Medium vaccination rate'
```

```python
df.loc[df['people_fully_vaccinated_per_hundred'] >= 70, 'Vax category'] = 'High vaccination rate'
```

```python
df['Vax category'].value_counts(dropna = False)
```

```python
# Create categorical plot 

sns.set(style = 'ticks')
sns.scatterplot(x = 'total_vaccinations', y = 'people_fully_vaccinated_per_hundred', hue = 'Vax category', data = df)
```
